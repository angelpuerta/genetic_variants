```python
data
```

```python
with_amino = data.dropna()
ones = with_amino[with_amino["CLASS"] == 1]
zeros =  with_amino[with_amino["CLASS"] == 0]
with_amino
```

```python
default_stratification =len(zeros)/len(with_amino)
default_stratification
```

```python
# ones_train = ones.sample(frac=0.8)
# zeros_train = zeros.sample(len(ones_train))
train = with_amino.sample(frac=0.8)
train
```

```python
## ones_test = ones.loc[set(ones.index) - set(ones_train.index)]
## zeros_test = zeros.loc[set(zeros.index)- set(zeros_train.index)].sample(round(len(ones_test)* (1/(1-default_stratification))))
test = with_amino.loc[set(with_amino.index) - set(train.index)]
test
```

```python
len(test[(test["CLASS"] == 1)]) / len(test[(test["CLASS"] == 0)])
```

```python
X_train = train.drop(columns="CLASS")
y_train = train["CLASS"]
X_test = test.drop(columns="CLASS")
y_test = test["CLASS"]
```

```python
hist_params = {
    'loss': ["auto", "binary_crossentropy"],
    'learning_rate': [0.1, 0.05, 0.2],
    'max_iter': [50, 100, 150],
    'min_samples_leaf': [15, 25, 50],
    'l2_regularization': [0, 0.1]
}
hist_search = RandomizedSearchCV(HistGradientBoostingClassifier(), params, error_score='raise')
hist_search.fit(X_train, y_train)
```

```python
hist_search.cv_results_
```

```python
hist_search.best_params_, hist_search.best_score_
```

```python
hist = hist_search.best_estimator_
f1_score(y_test, hist.predict(X_test)), accuracy_score(y_test, hist.predict(X_test)), accuracy_score(y_test, np.zeros(len(y_test)))
```

```python
forest_params = {
    'n_estimators': [50, 100, 200],
    'criterion': ["gini", "entropy"],
    'max_depth': [50, 100, 150],
}
search_forest = RandomizedSearchCV(RandomForestClassifier(), forest_params, error_score='raise')
search_forest.fit(X_train, y_train)
```

```python
search_forest.cv_results_
```

```python
forest = search_forest.best_estimator_
f1_score(y_test, forest.predict(X_test)), accuracy_score(y_test, forest.predict(X_test)), accuracy_score(y_test, np.zeros(len(y_test)))
```
